Bar Graph

```python
import matplotlib.pyplot as plt

"""
A Bar Graph is a type of graph that uses either horizontal or vertical bars to show comparisons among categories.
One axis shows the specific categories being compared, and the other represents a discrete value.
Bar graphs are used to compare things between different groups or to track changes over time. They are best for usage when the changes are larger.
color sets color of the bars, align sets the bar alignment, width sets the bar width, edgecolor sets the bar edge color, lw sets the edge width
"""

x = ["C++", "C", "Golang", "Python", "Java", "Kotlin", "JavaScript", "Assembly"]
y = [58, 66, 47, 88, 74, 44, 80, 20]
print(len(x), len(y))

plt.bar(x, y, color="#780000", align="center", width=0.6, edgecolor="red", lw=4)
# Set the font size for x-axis labels
plt.xlabel("Programming Languages", labelpad=10)  # Adjust label padding
plt.ylabel("Scores", labelpad=10)  # Adjust label padding
plt.xticks(fontsize=8)
plt.show()
```
