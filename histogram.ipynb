Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

"""
A Histogram is a graphical representation of the distribution of a dataset. 
It is a type of bar plot where the x-axis represents the bin ranges while the y-axis gives information about frequency of data points in those bins. 
The cumulative parameter is a boolean that allows to switch between a simple histogram and a cumulative histogram.
Cumulative histogram is a way to represent the frequency of observations that are less than or equal to a certain value.
The bins parameter tells us the number of bins that we want in our histogram.
The bins are usually specified as consecutive, non-overlapping intervals of a variable. They can be of different sizes and not need to be of equal width.
The bins parameter can also be a sequence of scalars representing the bin edges.
The histtype parameter allows us to choose the type of histogram that we want. The options are 'bar', 'barstacked', 'step', 'stepfilled'.
"""

ages = np.random.normal(23, 1.9, 10000)
print(ages)

plt.hist(ages, bins=[ages.min(), 19, 20, 21, 22, 23, 24, 25, 26, ages.max()], cumulative=True, histtype="step", color='b', linewidth=10)
plt.show()
plt.hist(ages, bins=30, cumulative=True, histtype="bar", color='r', linewidth=2)
plt.show()
```
