Plot Styling

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

"""
We can style our plots using the style library from the matplotlib package.
The style.use() method is used to set the style of the plot.
The style library contains a number of styles that can be used to style the plot.
Here we are using the dark_background style.
"""

X_data = np.random.random(800) * 700
Y_data = np.random.random(800) * 700

plt.scatter(X_data, Y_data, marker="8", s=80, alpha=0.8)
plt.title("Scatter plot", fontsize=24)
style.use("dark_background")
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

first = np.linspace(1, 20, 100)
second = np.linspace(10, 66, 130)
third = np.linspace(5, 80, 150)
fourth = np.linspace(23, 120, 180)
data = np.concatenate([first, second, third, fourth])


plt.boxplot(data)
style.use("grayscale")
plt.show() 
```
