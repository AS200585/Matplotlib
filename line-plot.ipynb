Line Graph

```python
import matplotlib.pyplot as plt

"""
A Line Graph  is a type of chart that displays information as a series of data points called 'markers' connected by straight line segments.
It is used to track changes over short and long periods of time. When smaller changes exist, line graphs are better to use than bar graphs.
Line graphs can also be used to compare changes over the same period of time for more than one group. 
lw sets the line width, linestyle sets the stypling of the line, marker is used to specify marker type for data points.
ms sets the marker size, mec sets marker edge color, mfc sets marker face color.
"""

years = [2006 + x for x in range(16)]
weight = [88, 86, 83, 78, 74, 70, 75, 80, 81, 79, 77, 75, 80, 85, 83, 80]
print(len(years), len(weight))

plt.plot(years, weight, c = "blue", lw = 4, linestyle = "--", marker="o", ms = 10, mec = "red", mfc = "green")
plt.show()
# "r--" is a shorthand for a red dashed line (where r stands for red and -- for dashed)
plt.plot(years, weight, "r--", marker="o", ms = 5, mec = "blue", mfc = "cyan")
plt.show()
```
