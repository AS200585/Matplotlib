Subplotting

```python
import numpy as np
import matplotlib.pyplot as plt

"""
Subplotting is a technique used to plot multiple graphs in a single figure. It is used to compare different graphs and analyze them.
In this example, we are plotting 4 graphs of Trignometric Functions in a single figure. 
We are using the subplot() function to create a 2x2 grid of graphs. This means that we will have 2 rows and 2 columns of graphs.
We can access each graph using the variable specified. Eg., We have specified axes as the variable. axes[0, 0] for first graph etc.
Functions used:
# set_title() - To set the title of the graph
# set_xlabel() - To set the label of x-axis
# set_ylabel() - To set the label of y-axis
# suptitle() - To set the title of the figure
"""

x = np.arange(500)
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sin Wave")
axes[0, 0].set_xlabel("x-axis")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Cosine Wave")
axes[0, 1].set_ylabel("y-axis")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("Tan Wave")
axes[1, 0].set_xlabel("x-axis")

axes[1, 1].plot(x, np.log(x))
axes[1, 1].set_title("Log Function")
axes[1, 1].set_ylabel("y-axis")

fig.suptitle("Trignometric Functions and their Graphs", fontsize=25)
plt.show()
```
