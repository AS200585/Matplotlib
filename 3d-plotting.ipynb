3-D Plotting 

```python
import numpy as np
import matplotlib.pyplot as plt

"""
We can create 3D scatter plots using the `scatter` method of the `Axes3D` object.
This can be done using plt.axes() method with the argument projection="3d".
This method takes three arguments, the x, y, and z coordinates of the data points.
"""
axs = plt.axes(projection="3d")

x = np.random.random(250)
y = np.random.random(250)
z = np.random.random(250)

axs.scatter(x, y, z)
axs.set_title("3D Scatter Plot")
axs.set_xlabel("X-axis")
axs.set_ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

axs = plt.axes(projection="3d")

x = np.arange(0, 300, 0.1)
y = np.sin(x)
z = np.cos(x)

axs.plot(x, y, z)
axs.set_title("3D Plot")
axs.set_xlabel("X-axis 2")
axs.set_ylabel("Y-axis 2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

"""
We can create 3D surface plots using the `plot_surface` method of the `Axes3D` object.
cmap argument is used to set the color map of the surface plot. Some of the color are - Spectral, viridis, plasma, inferno, magma, cividis.
view_init method is used to set the viewing angle of the plot.
azim argument is used to set the azimuthal viewing angle. Azimuthal viewing angle is the angle of the view around the z-axis.
elev argument is used to set the elevation viewing angle. Elevation viewing angle is the angle of the view above the x-y plane.
"""
axs = plt.axes(projection="3d")

x = np.arange(-1, 50, 0.1)
y = np.arange(-1, 50, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

axs.plot_surface(X, Y, Z, cmap="inferno")
axs.set_title("3D Surface Plot(Meshgrid)")
axs.set_xlabel("X-axis 3")
axs.set_ylabel("Y-axis 3")
axs.view_init(azim=0, elev=28)
plt.show()
```
