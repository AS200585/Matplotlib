Animations in Matplotlib

```python
import matplotlib.pyplot as plt # type: ignore
import random

"""
Another feature of matplotlib is the ability to animate plots. This is done by using the plt.pause() function. 
plt.pause() function pauses the program for a given amount of time. In the example below, we simulate flipping a coin 5 times and plot the results. 
The plt.pause(0.01) function pauses the program for 0.01 seconds. This allows us to see the plot update in real-time.
Since we are using Jupiter notebook, hence our animation will be displayed in the output cell in form of multiple images and not as a motion picture.
"""
heads_tails=[0, 0]

for _ in range(5):
    heads_tails[random.randint(0, 1)] += 1
    plt.bar(["Heads", "Tails"], heads_tails, color=["#283618", "#fb8500"])
    plt.pause(0.01)
plt.show()
```
