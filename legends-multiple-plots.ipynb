Legends and Multiple Plots

```python
import numpy as np
import matplotlib.pyplot as plt

"""
Legends are used to describe what each line represents. 
The legend() function is used to add a legend to the plot.
The legend() function takes an optional keyword argument loc that can be used to specify where in the plot the legend is to be drawn.
The loc argument can take the following values:
# 'best' (default): 0, 
# upper right': 1, 
# 'upper left': 2, 
# lower left': 3, 
# 'lower right': 4
# 'right': 5
# 'center left': 6
# 'center right': 7
# 'lower center': 8
# 'upper center': 9
# 'center': 10
We can also use the loc argument to specify the exact position of the legend in the plot by simply typing it.
We can also use the loc argument to specify the exact position of the legend as a tuple of two values. Eg., loc = (0.5, 0.5) for center placing.
"""

stock_a = [14, 23, 42, 131, 83, 157, 223, 182]
stock_b = [7, 13, 29, 33, 57, 99, 126, 189]
stock_c = [61, 89, 134, 156, 191, 234, 62]

plt.plot(stock_a, label='Company A')
plt.plot(stock_b, label='Company B')
plt.plot(stock_c, label='Company C')
plt.legend()
# plt.legend(loc = 'lower right')
# plt.legend(loc = 4)
# plt.legend(loc = (0.5, 0.5))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

votes = [131, 83, 157, 182]
party = ["A", "B", "C", "D"]

plt.pie(votes, labels=party)
plt.legend(labels=party)
plt.show()
```
