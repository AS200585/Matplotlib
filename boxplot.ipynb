Boxplot

```python
import numpy as np
import matplotlib.pyplot as plt

"""
A boxplot is a type of graph used in a standardized way for displaying the distribution of data based on a five number summary:-
minimum, first quartile(Q1), median, third quartile(Q3), and maximum. 
It can tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.
"""

first = np.linspace(1, 20, 100)
second = np.linspace(10, 66, 100)
third = np.linspace(5, 80, 100)
fourth = np.linspace(23, 120, 100)
data = np.concatenate([first, second, third, fourth])

plt.boxplot(data)
plt.show()
```
