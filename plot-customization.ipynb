Plot Customization

```python
import numpy as np
import matplotlib.pyplot as plt

"""
We can customize the plot using the following functions:
# xlabel() is used to set the label of the x-axis.
# ylabel() is used to set the label of the y-axis.
# title() is used to set the title of the plot.
# xticks() is used to set the ticks of the x-axis.
# yticks() is used to set the ticks of the y-axis.
# fontname is used to set the font of the text in the plot.
# fontsize is used to set the size of the text in the plot.
# There are many other functions that can be used to customize the plot. You can find them in the official documentation of Matplotlib.
"""

years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
incomes = [57, 56, 63, 62, 61, 71, 73, 74, 75]
income_ticks = list(range(51, 83, 2))

plt.plot(years, incomes)
plt.title("Monthly Income over the Years", fontsize=20, fontname="Comic Sans MS")
plt.xlabel("Years")
plt.ylabel("Income in Rupees(INR)")
plt.yticks(income_ticks, [f"INR {x}k" for x in income_ticks])
plt.show()
```
