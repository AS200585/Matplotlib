Multiple Figures

```python
import numpy as np
import matplotlib.pyplot as plt

"""
figure() function is used to create new figures which are independent of each other
The independent figures are identified by a number. The figure number is passed as an argument to the figure() function.
When no argument is passed to the figure() function, it creates a new figure with a new number.
This helps in creating multiple plots in the same program without needing to call the plt.show() argument multiple times.
"""

a1, a2 = np.random.random(500), np.random.random(500)
b1, b2 = np.arange(1000), np.arange(1000)

plt.figure(1)
plt.scatter(a1, a2)
plt.figure(2)
plt.plot(b1, b2)

plt.show()
```
