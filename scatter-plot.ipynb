Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

"""
A Scatter plot is a type of data visualization that shows the relationship between two numerical variables. 
Each member of the dataset gets plotted as a point whose x-y coordinates relates to its values for the two variables.
The scatter plot is a mainstay of statistical visualization. It is a collection of points placed using Cartesian coordinates.
The marker size is determined by value of s, which can be a scalar or a sequence of scalars.
The face color of the marker is determined by c, which can be a sequence of colors or a sequence of numbers.
The edgecolor of the marker is determined by edgecolors.
The transparency of the marker is determined by alpha.
The marker can be any of the following:- "." point, "," pixel, "o" circle, "v" triangle_down, "^" triangle_up, "<" triangle_left, ">" triangle_right
"1" tri_down, "2" tri_up, "3" tri_left, "4" tri_right,  "8" octagon,  "s" square
"""

X_data = np.random.random(500) * 500
Y_data = np.random.random(500) * 500

plt.scatter(X_data, Y_data, c="#000", marker="*", s=130, alpha=0.5)
plt.scatter(x = [200, 101, 420, 49], y =[420, 49, 101, 200], c="red", marker="v", edgecolors="maroon", s=180, alpha=0.9)
plt.show() 
```
