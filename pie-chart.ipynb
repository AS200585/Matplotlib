Pie Chart

```python
import matplotlib.pyplot as plt

"""
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.
In a pie chart, the arc length of each slice (and therefore its size) is proportional to the quantity it represents.
A pie chart is a good way to show the relationship of parts to a whole. It is a popular chart type in business and the media.
labels: A sequence of strings providing the labels for each wedge.
explode: A sequence of scalars, which will determine how far each wedge is offset from the center.
autopct: A string or function used to label the wedges with their numeric value. The label will be placed inside the wedge.
pctdistance: The ratio between the center of each pie slice and the start of the text generated by autopct.
startangle: The angle by which the start of the pie chart is rotated, counterclockwise from the x-axis.
shadow: A boolean value indicating whether to draw a shadow beneath the pie.
labeldistance: The radial distance at which the pie labels are drawn.
"""
languages = ['Python', 'Java', 'C', 'JavaScript', 'Swift', 'COBOL']
values = [65, 38, 26, 40, 46, 3]
explodes = [0.2, 0.1, 0, 0.2, 0.2, 0.6]

plt.pie(values, labels=languages, explode=explodes, autopct='%.4f%%', pctdistance=0.4)
plt.show()
plt.pie(values, labels=languages, explode=[0.2, 0.1, 0, 0.2, 0.2, 0.6], autopct='%.2f%%', pctdistance=0.8, startangle=90, shadow=True)
plt.show()
plt.pie(values, labels=languages, autopct='%.1f%%', pctdistance=0.6, startangle=30, labeldistance=0.8)
plt.show()
```
