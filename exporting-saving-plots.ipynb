Exporting, Saving and other functions on Plots

```python
import numpy as np
import matplotlib.pyplot as plt

# Using same program from subplotting file
"""
We can save the figure using savefig() method. This method takes the filename as the first argument and the format as the second argument.
The format can be png, jpg, pdf, svg, etc. The figure will be saved in the current working directory.
The figure can be saved in different resolutions using the dpi argument. The default value is 100.
transparent argument can be used to save the figure with a transparent background. The default value is False.
bbox_inches argument can be used to specify the bounding box in inches. The default value is tight.
pad_inches argument can be used to specify the padding in inches. The default value is 0.1.
Other values that can be used in bbox_inches are:- 
1. tight: This will adjust the bounding box to fit the figure.
2. medium: This will leave a little extra space around the figure.
3. loose: This will leave a lot of extra space around the figure.
4. None: This will leave no extra space around the figure.
"""
x = np.arange(500)
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sin Wave")
axes[0, 0].set_xlabel("x-axis")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Cosine Wave")
axes[0, 1].set_ylabel("y-axis")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("Tan Wave")
axes[1, 0].set_xlabel("x-axis")

axes[1, 1].plot(x, np.log(x))
axes[1, 1].set_title("Log Function")
axes[1, 1].set_ylabel("y-axis")

fig.suptitle("Trignometric Functions and their Graphs", fontsize=20)
# Ensure that savefig() is called before show() otherwise the figure saved will be blank or completely empty
# Saving the figure
plt.tight_layout() # This will ensure that the subplots do not overlap. Can be used with bbox_inches="tight" to remove extra white space
plt.savefig("trignometric-function.png", dpi=450, transparent=False, bbox_inches="tight", pad_inches=0.5)

plt.show()
```
